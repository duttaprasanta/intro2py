# Assignment 8
1. Take a number in *num*.
2. Write a function named *pos_neg_zero(num)*, which will return *'positive'*, *'negative'* or *'zero'* as output.
3. Write a function named *create_file_with_num(num,fname)* to create a file named *fname* and write *num* in the 1st line of it. Close the file.
4. Now write a function named *read_file_and_write_num_type(fname)* to read the file and then write the output of *pos_neg_zero(num)* in the next line of it. Close the file. It also returns the output.

> Use exception handling whereever possible.



```python
# @title Functions
def pos_neg_zero(num):
  match num:
    case num if num > 0:
      return 'positive'
    case num if num < 0:
      return 'negative'
    case num if num == 0:
      return 'zero'

def create_file_with_num(num, fname='checkType.txt'):
  try:
    with open(fname ,'w') as fs:
      fs.write(str(num))
  except Exception as e:
    print(f'Exception : {e}')

def read_file_and_write_num_type(fname='checkType.txt'):
  try:
    with open(fname,'r+') as fs:
      num = float(fs.readline().strip())
      numType = pos_neg_zero(num)
      fs.write(f'\n{numType}')
      return numType
  except Exception as e:
    print(f'Exception : {e}')
```

```python
num = 5
fname = 'checkType.txt'
create_file_with_num(num,fname)
read_file_and_write_num_type(fname)
```

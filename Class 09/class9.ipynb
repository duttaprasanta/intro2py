## Counter

```python
from collections import Counter
```

```python
print(Counter(['B','B','A','B','C','A','B','B','A','C']))

print(Counter({'A':3, 'B':5, 'C':2}))

print(Counter((Counter(A=3, B=5, C=2))))

print(Counter())
```

```python
counter = Counter({'A':1,'B':2, 'C':3})
counter.update({'A':1})
counter
```

```python
counter1 = Counter(A=3, B=5, C=2)
counter2 = Counter(A=1, B=2, C=3)
print(counter1 + counter2)
print(counter1 - counter2)
print(counter1 & counter2)
```

```python
from collections import Counter
my_counter = Counter('abracadabra')
print(my_counter.keys())
print(my_counter.values())
print(my_counter.items())
```

## Ordered Dict

Ordered Dict is a Dictionary in which the insertion order of an element is preserved.

```python
from collections import OrderedDict

od = OrderedDict()
od['A'] = 1
od['B'] = 2
od['C'] = 3
print(od)
```

```python
# @title Reverse a OrderedDict
from collections import OrderedDict

def reverse_ordered_dict(ordered_dict):
    return OrderedDict(reversed(ordered_dict.items()))

reverse_ordered_dict({1: 'a', 2: 'b', 3: 'c'})
```

```python
from collections import OrderedDict

my_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
last_item = my_dict.popitem(last=True)
a_val = my_dict.pop('a')

print(last_item)
print(a_val)
print(my_dict)
```

```python
from collections import OrderedDict

my_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])

print(my_dict)
# Move key 'a' to the end
my_dict.move_to_end('a')

# Move key 'c' to the beginning
my_dict.move_to_end('c', last=False)

my_dict
```

```python
# @title What is the output of this code
from collections import OrderedDict

# Create an ordered dictionary of key-value pairs
my_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])

# Add a new item to the end of the dictionary
my_dict['d'] = 4

my_dict.update([('e', 5), ('f', 6)])
my_dict.move_to_end('e', last=False)

# Iterate over the dictionary in the order in which items were added
for key, value in my_dict.items():
    print(key, value)
```

- Design a stack using OrderedDict. Write the functions for the stack as follows
  - push(stack,item)
  - pop(stack)
  - peek(stack)
  - is_empty(stack)
  - size(stack)
  - printStack(stack)

  Use the appropiate Exception Handling wherever required.



```python
from collections import OrderedDict

def push(stack, item):
  try:
    stack[len(stack) + 1] = item
  except Exception as e:
    print('Stack is full')

def pop(stack):
  try:
    return stack.popitem(last=True)[1]
  except Exception as e:
    print('Stack is empty')

def peek(stack):
  try:
    return stack[len(stack)]
  except Exception as e:
    print(e)

def is_empty(stack):
    return len(stack) == 0

def size(stack):
  return len(stack)

def printStack(stack):
  print('Stack : ',end='')
  for val in stack.values():
    print(val,end=' ')
  print('\n')
```

```python
stack = OrderedDict()
print('push(stack,1) => ')
push(stack,1)
printStack(stack)
print('push(stack,2) => ')
push(stack,2)
printStack(stack)
print('push(stack,3) => ')
push(stack,3)
printStack(stack)
print('pop(stack) => ',pop(stack))
printStack(stack)
print('peek(stack) => ',peek(stack))
printStack(stack)
print('is_empty(stack) => ',is_empty(stack),'\n')
print('size(stack) => ',size(stack),'\n')
print('pop(stack) => ',pop(stack))
printStack(stack)
print('pop(stack) => ',pop(stack))
printStack(stack)
print('pop(stack) => ',pop(stack))
printStack(stack)
```

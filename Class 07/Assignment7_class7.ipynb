# Assignment 7
Design your own *reduce* function named **myreduce(function, iterable, initializer=None)**, which works as original *reduce* function.





```python
def myreduce(function, iterable, initializer=None):
    it = iter(iterable)
    if initializer is None:
        value = next(it)
    else:
        value = initializer
    for element in it:
        value = function(value, element)
    return value

print(myreduce(lambda x,y:x+y, [1,2,3,4,5],5))

import functools
print(functools.reduce(lambda x,y:x+y, [1,2,3,4,5],5))
```

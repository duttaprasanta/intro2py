### Operators
---

* Arithmetic Operators
* Comparison Operators
* Logical Operators
* Bitwise Operators
* Assignment Operators
* Membership Operators

The precedence of Arithmetic Operators in Python is as follows:

1. P – Parentheses
2. E – Exponentiation
3. M – Multiplication (Multiplication and division have the same precedence)
4. D – Division
5. A – Addition (Addition and subtraction have the same precedence)
6. S – Subtraction

#### Arithmetic Operator

````python
p = 27
q = 3
print("p : ",p)
print("q : ",q)
print("p+q : ",p+q) # Addition
print("p-q : ",p-q) # Subtraction
print("p*q : ",p*q) # Multiplication
print("p/q : ",p/q) # Division
print("p//q : ",p//q) # Floor Division
print("p%q : ",p%q) # Modulus
print("p**q : ",p**q) # Exponentiation
````

#### Comparison Operator

````python
a=5; b=6
print("a == b : ",a == b)
print("a != b : ",a != b)
print("a > b : ",a > b)
print("a < b : ",a < b)
print("a >= b : ",a >= b)
print("a <= b : ",a <= b)
print("1 < a <10 : ", 1 < a <10) # Chaining comparison Operator
````

#### Logical Operators

````python
a = True
b = False
print("a : ",a)
print("b : ",b)
print("a and b : ",a and b)
print("a or b : ",a or b)
print("not a : ",not a)
````

#### Assignment Operator

````python
a = 10
a += 10 # a = a + 10
a -= 10 # a = a - 10
a *= 10 # a = a * 10
a /= 10 # a = a / 10
a %= 10 # a = a % 1
a **= 10 # a = a ** 10

b = [1, 2, 3, 4, 5]

# walrus operator
while(x := len(b) > 2):
    b.pop()
    print(x)
````

#### Bitwise Operator

````python
a = 10
b = 20
print("a : ",a)
print("b : ",b)
print("Binary of a : ", bin(a))
print("Binary of b : ", bin(b))

print("\nBinary of (a & b)", bin(a & b))
print("a & b : ",a & b) # Bitwise And

print("\nBinary of (a | b)", bin(a | b))
print("a | b : ",a | b) # Bitwise Or

print("\nBinary of (a ^ b)", bin(a ^ b))
print("a ^ b : ",a ^ b) # Bitwise Xor

print("\nBinary of (~a)", bin(~a))
print("~a : ",~a) # Bitwise Not

print("\nBinary of (~b)", bin(~b))
print("~b : ",~b) # Bitwise

print("\nBinary of (a << 2)", bin(a << 2))
print("a << 2 : ",a << 2) # Bitwise Left Shift

print("\nBinary of (a >> 2)", bin(a >> 2))
print("a >> 2 : ",a >> 2) # Bitwise Right Shift
````

#### String Slicing
One thing to keep in mind while using this method is that the string returned after slicing includes the character at the start index but not the character at the last index.

<img src='https://media.geeksforgeeks.org/wp-content/cdn-uploads/20200204160843/strings.jpg'>

Syntax
```python
var[start=0:end=len:step=1]
```

````python
str8 = '0123456789'
str8[2:10]
````

````python
str8 = '0123456789'
str8[2:-2]
````

````python
str8 = '0123456789'
str8[0::2]
````

##### Printing using escape characters

````python
String = "This is \x69\x6e\x74\x72\x6f\x32\x70\x79 \x63\x6c\x61\x73\x73"
print(String)
````

````python
# @title Printing raw string

String = r"This is \x69\x6e\x74\x72\x6f\x32\x70\x79 \x63\x6c\x61\x73\x73"
print(String)
````

#### More formatting options

````python
print("Binary of {0} is {0:b}".format(8))
print("Octal of {0} is {0:o}".format(8))
print("Hexadecimal of {0} is {0:x}".format(8))
````

````python
print("Exponential of {0} is {0:e}".format(123.4569))
````

````python
print("The value of {0}/{1} is {2:.2f}".format(1,6,1/6))
````

````python
print("{:>10} | {:^10} | {:<10}".format('Roll', 'Name', 'Marks'))
````

Take a string as an input from user. Trim the spaces (both sides). Then reverse the string using String Slicing only.

Note:
1. Don't use any inbuild function
2. Don't use loops and conditional statements

```python
string = input("Enter a string: ")
string = string.strip()
string = string[::-1]
print(f'The reversed string is : {string}')
```

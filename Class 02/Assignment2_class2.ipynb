# Assignment 2:
Take a character as an input. Print the ASCII value of that character in two different ways.

```python
# @title Method1
char = input("Enter a character : ")
print(f'The ASCII value of {char} is  {ord(char)}')
```

```python
# @title Method2
char = input("Enter a character : ")
print(f'The ASCII value of {char} is {char.encode()[0]}')
```

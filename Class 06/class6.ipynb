## Loop
Python loop (for & while) supports the following functions -

* break
* continue
* else
* pass


### For loop
```python
for var in iterable:
  do something
```

````python
for i in range(10):
  print(i, end=' ')
````

````python
for i in range(1, 11):
  print(i, end=' ')
````

````python
for i in range(1, 11, 2):
  print(i, end=' ')
````

````python
for i in range(10, 0, -1):
  print(i, end=' ')
````

````python
for x in 'This is python class':
  print(x, end=' ')
````

````python
for i,j in enumerate(['eat','work','sleep','repeat']):
  print(i,j)
````

````python
for i in range(2):
  for j in range(3):
    print(i,j)
````

````python
numbers_10 = [i+1 for i in range(10)]
numbers_10
````

````python
list4 = [[1,2,3],[4,5,6],[7,8,9]]
print([j for i in list4 for j in i])
````

````python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
for i,j in zip(list1,list2):
  print(i+j, end=' ')
````

````python
for i in range(100):
  if i ** 0.5 != int(i ** 0.5):
    continue
  print(i, end=' ')
````

````python
for i in range(100):
  if i>20:
    break
  print(i, end=' ')
````

````python
for i in (1,2,3,4,5):
  pass
````

````python
for i in range(10):
  print(i, end=' ')
else:
  print('Done')
````

````python
for i in range(10):
  if i>5:
    print('Break')
    break
  print(i, end=' ')
else:
  print('Done')
````

### While Loop
Supports all the features of for. Only the syntax is different.

````python
i=0
while i<10:
  print(i, end=' ')
  i+=1
````

````python
i=0
while True:
  if i>10:
    break
  print(i, end=' ')
  i+=1
````

````python
list3 = [1,2,3,4,5]
while list3:
  print(list3.pop(), end=' ')
````

````python
#dictionary comprehension
print({i:j**2 for i,j in enumerate(range(10))})
````

````python
print(dict.fromkeys(range(5), True))
````

### Scope of a variable

````python
global_var = 'Programming'

def scope_example():
  local_var = 'Introduction to Python'
  print(f'Inside function - {local_var}')
  print(f'Inside function - {global_var}')

scope_example()
print(f"Outside function - {global_var}")
````

````python
var = 'Programming'

def scope_example():
  print(f'Inside function - {var}')

scope_example()
print(f"Outside function - {var}")
````

````python
var = 'Programming'

def scope_example():
  var = 'Introduction to Python'
  print(f'Inside function - {var}')

scope_example()
print(f"Outside function - {var}")
````

````python
var = 'Programming'

def scope_example():
  global var
  var = 'Introduction to Python'
  print(f'Inside function - {var}')

scope_example()
print(f"Outside function - {var}")
````

### Return statement in function

````python
def return_example():
  print('Hello world')

ret_val = return_example()
print(f'The returned value is : {ret_val}')
````

````python
def return_example():
  print('Hello world')
  return 'Hello World'

ret_val = return_example()
print(f'The returned value is : {ret_val}')
````

#### Returning Multiple values
We can return multiple values in terms of iterables

````python
def multiple_return():
  return 1,2,3

a,b,c = multiple_return()
print("The returned values are : ",a,b,c)
````

### Closures

````python
def outer_func(x):
  def inner_func(y):
    return x+y
  return inner_func

add_5 = outer_func(5)
print(add_5(10))
````

**What is the utility of it?**


````python
def power_function(exponent):
    def calculate(x):
        return x ** exponent
    return calculate

square = power_function(2)
cube = power_function(3)

print(square(4))
print(cube(2))
````

````python
def mult(x):
  def multiply(y):
    return x*y
  return multiply

double = mult(2)
triple = mult(3)

print(double(5))
print(triple(6))
````

### Arguments and Keyword-Arguments

````python
def args_example(*args, **kwargs):
  print(f'Arguments : {args}')
  print(f'Keyword arguments : {kwargs}')
````

````python
args_example(1,2,3,4)
````

````python
args_example(a=1,b=2,c=3)
````

````python
args_example(1,2,3,a=1,b=2,c=3) # Kwargs should be specified after the args
````

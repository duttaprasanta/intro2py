## Assignment 6
Write four functions for printing the following diagrams for a given n, where n the number of '*' in the base/ height of the triangle. Take n as a input from the user.

```
     *
    **
   ***
  ****
 *****
******

```

```
*
**
***
****
*****
******
```
```
******
*****
****
***
**
*
```
```
******
 *****
  ****
   ***
    **
     *
```




````python
def upper_right_triangle(n):
  for i in range(1, n+1):
    print('*' * i)

def lower_right_triangle(n):
  for i in range(n, 0, -1):
    print('*' * i)

def upper_left_triangle(n):
  for i in range(1, n+1):
    print(' ' * (n-i) + '*' * i)

def lower_left_triangle(n):
  for i in range(n, 0, -1):
    print(' ' * (n-i) + '*' * i)

def print_using_triangle(n):
  upper_left_triangle(n)
  print()
  upper_right_triangle(n)
  print()
  lower_left_triangle(n)
  print()
  lower_right_triangle(n)

n = int(input("Enter the value of n : "))
print_using_triangle(n)
````

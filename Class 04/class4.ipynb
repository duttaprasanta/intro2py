## Set
A Python set cannot have mutable elements like a list or dictionary. It contains the unique values.

### Initialisation

````python
# Empty set
set1 = set()

# Set with elements
set2 = {1, 2, 3}

# Set from a list
set3 = set([1, 2, 3, 4, 5])

# Set with mixed data types
set4 = {1, 'hello', 3.14, (1, 2)}

# Set from a string (unique characters)
set5 = set('hello')

# Set with the use of dictionary
set6 = set({'a':1,'b':2,'c':3})
````

### Typecasting

````python
# Typecasting in Set
my_list = [1, 2, 3, 3, 4, 5]
my_set = set(my_list)  # Convert a list to a set, removing duplicates
print(my_set)  # Output: {1, 2, 3, 4, 5}

my_tuple = (1, 2, 3, 3, 4, 5)
my_set = set(my_tuple)  # Convert a tuple to a set, removing duplicates
print(my_set)  # Output: {1, 2, 3, 4, 5}

my_string = "hello"
my_set = set(my_string)  # Convert a string to a set, containing unique characters
print(my_set)  # Output: {'h', 'e', 'l', 'o'}

my_set = {1, 2, 3}
my_list = list(my_set)  # Convert a set to a list
print(my_list)  # Output: [1, 2, 3] (Order may vary)

my_set = {1, 2, 3}
my_tuple = tuple(my_set)  # Convert a set to a tuple
print(my_tuple)  # Output: (1, 2, 3) (Order may vary)
````

### Set Operations

````python
# Set Operations

set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Union
union_set = set1 | set2
print("Union:", union_set)  # Output: {1, 2, 3, 4, 5}

# Intersection
intersection_set = set1 & set2
print("Intersection:", intersection_set)  # Output: {3}

# Difference
difference_set = set1 - set2
print("Difference (set1 - set2):", difference_set)  # Output: {1, 2}

difference_set = set2 - set1
print("Difference (set2 - set1):", difference_set)  # Output: {4, 5}

# Symmetric Difference
symmetric_difference_set = set1 ^ set2
print("Symmetric Difference:", symmetric_difference_set)  # Output: {1, 2, 4, 5}


# Using methods
union_set = set1.union(set2)
print("Union (using method):", union_set)

intersection_set = set1.intersection(set2)
print("Intersection (using method):", intersection_set)

difference_set = set1.difference(set2)
print("Difference (using method):", difference_set)

symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference (using method):", symmetric_difference_set)

# Checking Subset and Superset
set3 = {1, 2}
print("Is set3 a subset of set1?", set3.issubset(set1))  # Output: True
print("Is set1 a superset of set3?", set1.issuperset(set3))  # Output: True
````

1. Create a list of dictionaries where each dictionary contains information about a book. The dictionary should have keys: "title", "author", and "year". Populate this list with three books.
2. Create a set of all unique authors from the list.
3. Print the list of books and the set of unique authors.

Note : Only use the datastructures and functions that we have seen so far.

```python
books = [{'title':'title1','author':'author1','year':'2024'},
 {'title':'title2','author':'author1','year':'2021'},
  {'title':'title3','author':'author2','year':'1992'}]
```

```python
unique_authors = set({books[0]['author'],books[1]['author'],books[2]['author']})
```

```python
[books[0]['title'],books[1]['title'],books[2]['title']], list(unique_authors)
```

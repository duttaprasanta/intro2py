##### F-string

````python
a, b  = 50, 100
print(f"The value of a is {a} and the value of b is {b}")
````

````python
import datetime
print(datetime.datetime.today())
print(f'date: {datetime.datetime.today():%Y-%m-%d}')
````

Note: Backslash Cannot be used in format string directly


```python
print(f"newline: {ord('\n')}") # Will produce error
```

Workaround: Use \ contained in a variable

````python
print(f"newline: {ord('\n')}")
````

````python
x = ord('\n')
print(f"newline: {x}")
````

Note: While working with dictionaries, we have to make sure that if we are using double quotes (“) with the f-string then we have to use single quote (‘) for keys inside the f-string in Python and vice-versa. Otherwise, it will throw a syntax error.

##### Formatting using %

````python
a,b = 50, 100
print("The value of a is %d and the value of b is %d" % (a, b))
````

``` python
 %[width][.precision]type
```

Types

* d = Decimal integer
* b = Binary format
* o = octal format
* x or X = Hexadecimal format
* e or E = Exponential notation
* f or F = Floating-point decimal
* g or G = General format
* c = Single Character
* r = String format(using repr())
* s = String Format(using str()))





````python
print("%5.2f"%(15.52563))
````

````python
print("%5d"%5.25356)
````

````python
print("%.2f"%5.25356)
````

````python
print("%9o"%52)
````

````python
print("%10.3E" % (356.08977))
````

#### Using Format

````python
roll = 25
name = "Prasanta"
#print(f"Roll: {roll} Name: {name}")
print("Roll: {} Name: {}".format(roll, name))
````

````python
print("Roll: {1} Name: {0}".format(name, roll))
````

````python
print("Roll: {y} Name: {x}".format(x=name, y=roll))
````

````python
x = 123.56
print("Roll: {0:4d} Name: {1:.1f}".format(roll, x))
````

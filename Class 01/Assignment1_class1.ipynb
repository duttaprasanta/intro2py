# Assignment 1
Print three different lines of text using a single print function in 3 different ways. The texts are as follows -

1. Python is fun.
2. Python is versatile.
3. Python is powerful.

## Solution 1

```python
print("Python is fun.", "Python is versatile.", "Python is powerful.", sep='\n')
```

## Solution 2

```python
print("Python is fun. \nPython is versatile. \nPython is powerful.")
```

## Solution 3

```python
print("""Python is fun.
Python is versatile.
Python is powerful.
""")
```

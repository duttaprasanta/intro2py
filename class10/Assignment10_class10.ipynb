# Assignment 10
Given a list of elements. Sort the elements in decending order using a heap.




```python
import heapq

def heapSort(l):
  heapq.heapify(l)
  return list(reversed([heapq.heappop(l) for i in range(len(l))]))

heapSort([1,3,5,7,9,2,4,6,8,0])
```
